```python
#First we get the directrory and strip off double quotes at the begining and end
_directory = input("Paste file directory: ").strip('\"')

#prints the path without the double quotes
print(f"This is the path entered {_directory}") 

#replace the back slash with Forward slash
_directory = _directory.replace("\\","/") 

#Open and read list all entries in the directory

# Using pathlib
from pathlib import Path
import pandas as pd
myPath = Path(_directory)

Folder_list = []
Files_list = []
for item in myPath.iterdir():
    if item.is_dir():
        Folder_list.append(item.name)
    else:
        Files_list.append(item.name)

#Equalize file_list and Folder_list using an empty string (" ") to ensure error free DataFrame creation subsequentli
if len(Files_list) != len(Folder_list):
     list_lenth_diff = max(len(Files_list),len(Folder_list)) - min(len(Files_list),len(Folder_list))
     if min(len(Files_list),len(Folder_list)) == len(Folder_list):
         for i in range(list_lenth_diff):
             Folder_list.append(" ")
     else:
         for i in range(list_lenth_diff):
             Files_list.append(" ")

         
df = pd.DataFrame({
    "FOLDERS": Folder_list,
    "FILES" : Files_list
})

print("Here are the list of Folders and Files in the Directory you entered \n")
print(df)

print("If you want to move and organize files, \n Press ENTER KEY to continue \n N.B: This file should not be in the directory you want to Organizes")
input()

print("\n")
#Determine the type of Each entry in the directry

import magic
import shutil

ms = magic.Magic(mime=True)

for item in myPath.iterdir(): #Iterates through the folder
    if item.is_file():
        try:
            file_type = ms.from_file(str(item)) #This gets the file type e.g. [image/jpeg], [application/pdf]
            print(f"File: {item.name}")
            print(f" - Actual Content Type: {file_type}")
#
#Creates a new folder and move the file into it

            file_Category = file_type.split("/")[0] #This separates the file category to from the extension ie [image/png] returns image

            #Creates the destination folder
            dest_folder = myPath / file_Category    
            dest_folder.mkdir(exist_ok=True)  #creates a new folder or ignores if folder exist

            #Check if file with same name exist in the folder else move the file to the folder
            dest_path = dest_folder/item.name

            if dest_path.exists():
                print(f"Skipping {item.name} because it already exists in {file_Category}")
            else:
                shutil.move(str(item),str(dest_path))
                print(f"Moved {item.name} to {file_Category}/")

        except Exception as e:
            print(f" Error Moving {item.name}: {e}")
        
       
            
```
